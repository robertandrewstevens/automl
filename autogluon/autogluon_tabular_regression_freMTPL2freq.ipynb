```python
from autogluon.tabular import TabularDataset, TabularPredictor
# from sklearn.model_selection import train_test_split
```

## Data

```python
# TODO: Download the suggested data from Zindi
# https://zindi.africa/competitions/ai-tunisia-hack-5-predictive-analytics-challenge-2
# data = TabularDataset("data/Train.csv")
# data = TabularDataset("Train_zindi.csv")
train_data = TabularDataset("freMTPL2freq_dataset_train.csv")
train_data.head()
```

```python
# train, val = train_test_split(data, test_size=0.3, shuffle=True)
```

```python
# test = TabularDataset("data/Test.csv")
# test = TabularDataset("Test_zindi.csv")
test_data = TabularDataset("freMTPL2freq_dataset_test.csv")
test_data.head()
```

## Modeling

```python
# LABEL = 'target'
LABEL = "ClaimNb"
# SAVE_PATH = 'AutoGluonModels_exercise'
SAVE_PATH = 'AutoGluonModels_freMTPL2freq'
EVAL_METRIC = 'root_mean_squared_error'
PROBLEM_TYPE = 'regression'
```

```python
# TODO: Fit the Tabular Predictor and enable bagging
predictor = TabularPredictor(
    label=LABEL, 
    problem_type=PROBLEM_TYPE, 
    path=SAVE_PATH, 
    eval_metric=EVAL_METRIC
)
```

```python
predictor.fit(train_data, presets=["best_quality"], time_limit=300)
```

```python
predictor.leaderboard(test_data, extra_info=False, silent=True)
```

```python
# TODO: Make predictions on the test set
predictions = predictor.predict(test_data)
predictions.head()
```

## Hyperparameter Tuning

```python
predictor_hpt = TabularPredictor(
    label=LABEL, 
    problem_type=PROBLEM_TYPE, 
    path=SAVE_PATH + "_hpt", 
    eval_metric=EVAL_METRIC
).fit(
    train_data, 
    presets=['best_quality'], 
    time_limit=3600, 
    hyperparameters={
        'RF': [{'max_depth': 6}],
        'GBM': [{'max_depth': 6}],
        'XGB': [{'max_depth': 12}]
    }
)
```

```python
predictor_hpt.leaderboard(test_data, extra_info=False, silent=True)
```

```python
predictions = predictor.predict(test_data)
predictions.head()
```
